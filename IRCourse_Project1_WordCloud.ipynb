```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Using the CSVs from WordCount, create a wordcloud
df = pd.read_csv('P1_WordCount.csv', delimiter='\t')
df = df.sort_values(by=['Count'], ascending=False)
word_list = df.head(20).copy()
word_list
```

```python
df = pd.read_csv('P1_WordCount_StopWords.csv', delimiter='\t')
df = df.sort_values(by=['Count'], ascending=False)
df = df.drop([17, 20]) 
word_list_stopwords = df.head(20).copy()
word_list_stopwords
```

```python
# Convert DFs to dictionaries of word : count
data = dict(zip(word_list["Word"].tolist(), word_list["Count"].tolist()))
data_stop = dict(zip(word_list_stopwords["Word"].tolist(), word_list_stopwords["Count"].tolist()))
```

```python
wc = WordCloud(background_color="white", max_words=20).generate_from_frequencies(data) 
plt.figure(figsize=(15,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(background_color="white", max_words=20).generate_from_frequencies(data_stop) 
plt.figure(figsize=(15,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
